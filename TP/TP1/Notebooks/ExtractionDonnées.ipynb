# Extraction à partir d'un fichier CSV

### 1. Charger le contenu du fichier précédent (csv) et afficher le Dataframe résultat (df1_csv)

```python
file_path = "../Data/logdata1.csv"
df1_csv = pd.read_csv(file_path)
df1_csv
```

### 2. Charger le contenu du fichier (csv)tout en changeant les noms de colonnes et le stocker dans un DataFrame (df2_csv)

```python
df2_csv = pd.read_csv(file_path, names = ["bill", "tips", "gender", "fumeur", "jour", "temps", "taille"])
df2_csv
```

### 3. Charger le contenu du fichier (csv) dans un dataftrame (df3_csv) tout en sautant les 2 premières lignes du fichier

```python
df3_csv = pd.read_csv(file_path, skiprows = [0,1])
df3_csv
```

### 4. Charger le contenu du fichier (csv) dans un dataftrame (df4_csv) tout en lisant les 100 premières lignes du fichier

```python
df4_csv = pd.read_csv(file_path, nrows = 100)
df4_csv
```

### 5. Charger le contenu des 2 premières colonnes du fichier (csv) dans un dataftrame (df5_csv)

```python
df5_csv = pd.read_csv(file_path, usecols = [0,1])
df5_csv
```

### 6. Charger le contenu des 2 colonnes au choix du fichier (csv) dans un dataftrame (df6_csv)

```python
df6_csv = pd.read_csv(file_path, usecols = ["smoker", "sex"])
df6_csv
```
